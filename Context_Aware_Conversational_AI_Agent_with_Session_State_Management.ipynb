## 🧠 Day 6: Autonomous AI Agent with Memory & State Maintenance
* **Core Concepts:** Persistent State management, Python Lists (`.append()`), Conversation history tracking.
* **Purpose:** Builds an engine that retains contextual memory across multiple turns of chat, mimicking production-grade conversational agents.
* **Status:** Complete ✅


```python
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

print("AI agent with Memory Loaded ----")
print("Type 'exit' to end the conversation.\n")

conversation_history = []

system_instructions = "You are a helpful AI assistant. Always review the past conversation history provided to maintain context."

while True:
  user_input = input("You: ")

  if user_input.lower() == 'exit':
    print("Agemt: Memory cleared. Goodbye!")
    break

  if not user_input.strip():
    continue

  conversation_history.append(f"User: {user_input}")
  history_context = "\n".join(conversation_history)

  prompt = f"""
  {system_instructions}

  --- CONVERSATION HISTORY STARTS ---
  {history_context}
  --- CONVERSATION HISTORY ENDS ---

  Provide the next logical response as the AI assistant.
  """

  print("Agent is thinkng...")

  response = client.models.generate_content(
      model = 'gemini-2.5-flash',
      contents = prompt,
  )

  ai_response = response.text.strip()

  print(f'\nAgent: {ai_response}\n')

  conversation_history.append(f'Agent: {ai_response}')
```
