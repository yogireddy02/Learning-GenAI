## 📝 Day 3: Automated News Summarizer
* **Core Concepts:** Python File Handling (`open()`, `write()`), Context Managers (`with` block).
* **Purpose:** Generates 3-bullet point executive text summaries and commits them directly to a local `.txt` file.
* **Status:** Complete ✅


```python
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key = api_key)

print("--- AI News Summarizer Loaded ---")

long_article = """
Bengaluru is experiencing a massive boom in Global Capability Centers (GCCs), with
over 50 new centers established in the last year alone. Companies like GCCs are
shifting focus from basic IT operations to advanced Artificial Intelligence and
Machine Learning research. This shift has triggered an intense talent war for
engineers skilled in Agentic AI and large language model orchestration. According
to industry reports, entry-level salaries for AI specialists have surged by 40%
compared to traditional software testing roles. Industry experts predict that by 2027,
India will host over 2,000 GCCs, making it the undisputed tech capital for autonomous
enterprise automation.
"""

instructions = f"""
You are an expert news editor.
Summarize the following article into exactly three concise bullet points.
Focus only on the most critical takeaways.

Article:
{long_article}
"""

print("AI is summarizing the article...")

response = client.models.generate_content(
    model = 'gemini-2.5-flash',
    contents = instructions,
)

summary_text = response.text

print("\n--- AI SUMMARY ---")
print(summary_text)


file_name = "news_summary.txt"

with open(file_name, 'w') as file:
  file.write("==== BUSINESS NEWS SUMMARY ====\n")
  file.write(summary_text)

print(f"\n[SUCCESS] File successfully saved as '{file_name}'!")
```
