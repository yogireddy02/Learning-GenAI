## 📊 Day 2: AI Expense Categorizer with Structured Output
* **Core Concepts:** JSON schema enforcement, Python Dictionaries, Error Handling (`try/except`).
* **Purpose:** Forces the AI to return raw mathematical data structures instead of conversational text.
* **Status:** Complete ✅


````python
import json
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key= api_key)

print("---- AI Expense Categorizer Loaded ---")

raw_expenses = "Yesterday I paid ₹450 for a Zomato order, then took an auto for ₹120, and bought a movie ticket for ₹350."

instructions = f"""
You are a data extraction assistant.
Analyze the following sentence and extract the expenses.
Group them into logical categories like Food, Transport, Entertainment, Utilities, or Shopping.

Return the data ONLY as a valid JSON object where the keys are categories and the values are integers (numbers).
Do not include any conversational text, markdown formatting, or ```json blocks.

Sentence = {raw_expenses}
"""

print("AI is extracting data....")


response = client.models.generate_content(
    model = 'gemini-2.5-flash',
    contents = instructions,
)

try:
  expense_dict = json.loads(response.text.strip())

  print("\n--- EXTRACTED DATA SUCCESS ---")
  print("Python Data Type:", type(expense_dict))
  print("Final Dictionary:", expense_dict)

  total_spent = sum(expense_dict.values())
  print(f"Total Amount Spent: ₹{total_spent}")

except Exception as e:
  print("\n--- ERROR ---")
  print("The AI didn't return perfect data format. Raw response was:")
  print(response.text)
````
