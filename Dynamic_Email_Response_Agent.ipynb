```python
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

customer_email = """
Hi, I ordered a smartphone from your website three days ago.
The tracking page hasn't updated at all, and I need this phone before my trip tomorrow!
This is incredibly frustrating. Please cancel my order and refund my money immediately.
"""

instructions = f"""
You are an expert customer support assistant.
Analyze the following customer email. Perform two tasks:
1. Identify the tone or emotion (e.g, Angry, Happy, Anxious).
2. Write a polite, empathetic, and professional reply solving their issue.

Customer email:
{customer_email}
"""

print("Ai is analyzing the email and drafting a response")
response = client.models.generate_content(
    model = 'gemini-2.5-flash',
    contents = instructions,
)

print("\n--- AI ANALYSIS & REPLY ---")
print(response.text)
```

```python
from google import genai
from google.colab import userdata

api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

print("--- AI Support Assistant Loaded ---")
print("Type 'exit' at any time to close the program.\n")

while True:
  customer_email = input("paste the customer email here: ")

  if customer_email.lower() == "exit":
    print("Closing the assistant. Goodbye!")
    break

  if not customer_email.strip():
    print("Please enter some text!\n")
    continue

  instructions = f"""
  You are an expert customer support assistant.
  Identify the tone of this email and write a professional reply.

  Customer Email:
  {customer_email}
  """

  print("\nAI is thinking...")

  response = client.models.generate_content(
      model = 'gemini-2.5-flash',
      contents = instructions,
  )

  print("\n======================================")
  print(response.text)
  print("======================================\n")
```
